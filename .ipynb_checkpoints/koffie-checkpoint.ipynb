# Differentiequotiënt: Koffie

Het afkoelen van een kopje koffie hangt af van de temperatuur van de koffie bij het inschenken en van de kamertemperatuur. 
Ook de vorm en het materiaal waarvan het kopje is gemaakt, heeft invloed. 

De formule $T(t) = 20 + 70 . 0,82^t$ geeft de temperatuur van de koffie.
                                 

\begin{equation}
\Delta temp=\frac{T(t1) - T(t0)}{t1 - t0} \frac{°C}{min}
\end{equation}

```python
temp_omgeving = 20
koffie_temp_0 = 90
d_koffie = koffie_temp_0 - temp_omgeving

# begintijd in minuten
t0 = 0

def T(t):
    ''' t is de delta tijd in minuten '''
    afkoelings_coeff = 0.82**t
    
    return temp_omgeving + d_koffie * afkoelings_coeff


def delta_T(t0, t1):
    return (T(t1) - T(t0))/(t1-t0)

```

na 5 minuten

```python
tijd = list(range(0, 65, 5)) # van 0 to tot 60 minuten, in stappen van 5 minuten

# onderstaand is een 'list comprehension'
temp_daling = [T(t) for t in tijd] # voor elke tijd in de bovenstaande lijst, bereken de temp. Alternatief voor wat we met 'd_temp' doen.


t_diff = []
d_temp = []

for i, d_tijd in enumerate(tijd):  # enumerate geeft index en waarde voor elke waarde in de lijst
    if i == 0:  # begin, nog geen tijd gehad om af te koelen
        t_diff.append(0)
        d_temp.append(koffie_temp_0)
    else:
        d_coeff = delta_T(tijd[i-1], d_tijd)  
        t_diff.append(d_coeff)
        d_temp.append(d_temp[i-1] + (d_tijd - tijd[i-1])*d_coeff) 
                      
```

```python
import pandas as pd
row_names = ['coeff', 'temp']
df = pd.DataFrame([t_diff, d_temp], index=row_names, columns=tijd)
df
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
plt.plot(tijd, t_diff, marker='o') # 'o' marks each point on the plot
plt.title("koffie afkoeling: diff coeff")
plt.xlabel("tijd (min)")
plt.ylabel("°C/min")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(tijd, d_temp, marker='o') # 'o' marks each point on the plot
plt.title("koffie afkoeling: temperatuur")
plt.xlabel("tijd (min)")
plt.ylabel("°C")
plt.grid(True)
plt.show()
```
